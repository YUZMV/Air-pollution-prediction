### Introduction

La pollution de l’air est l’un des principaux facteurs de risque environnemental pour la santé humaine. Selon l’OMS, elle provoque chaque année plusieurs millions de décès prématurés dans le monde.  
La capacité à **prédire avec précision les niveaux de pollution** (PM2.5, PM10, NO₂, O₃, AQI, etc.) est donc cruciale pour :
- Anticiper les pics de pollution,
- Alerter les populations vulnérables,
- Aider les autorités à prendre des mesures correctives (circulation alternée, restrictions industrielles, etc.).

Ce projet s’inscrit dans cette démarche en développant un modèle de **régression** capable de prédire un indice de pollution (variable cible continue) à partir de 5 variables explicatives numériques anonymisées.</br></br>

### Problématique

Étant donné un jeu de données contenant :
- 1600 échantillons d’entraînement (avec cible)
- 400 échantillons de test (sans cible)
- 5 caractéristiques numériques (normalisées, sans nom explicite)

**Comment prédire de manière fiable et précise la valeur de la cible (niveau de pollution) sur de nouvelles observations ?**

Il s’agit d’un problème classique de **régression supervisée**.</br></br>

### Objectifs du projet

1. Explorer et comprendre la structure des données  
2. Identifier les relations entre les variables explicatives et la cible  
3. Construire, entraîner et évaluer plusieurs modèles de régression  
4. Sélectionner le modèle le plus performant (critères : R², MSE, robustesse)  
5. Générer les prédictions sur le jeu de test  
6. Proposer des axes d’amélioration et de déploiement futur
